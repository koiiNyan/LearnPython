```python
import pandas as pd
```

```python
export_2015 = pd.read_excel(open('2015.xlsx', 'rb'), sheet_name=1)
```

```python
export_2015.head()
```

```python
export_2016 = pd.read_excel(open('2016.xlsx', 'rb'), sheet_name=1)
```

```python
export_2016.head()
```

```python
export = pd.concat([export_2015, export_2016])
```

```python
export.sample(10)
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
export.pivot_table(index=['Year'], values='Export (US$ Thousand)', aggfunc='sum')
```

```python
export.pivot_table(index=['Year','Partner Name'], values='Export (US$ Thousand)', aggfunc='sum').sort_values('Export (US$ Thousand)', ascending=False)
```

```python
exclude = ['Europe & Central Asia', 'East Asia & Pacific', 'World', 'Unspecified', 'Middle East & North Africa	']
export_filtered = export[-export['Partner Name'].isin(exclude)]

export_filtered.pivot_table(index=['Year','Partner Name'], values='Export (US$ Thousand)', aggfunc='sum').sort_values('Export (US$ Thousand)', ascending=False)
```
